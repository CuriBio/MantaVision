```python
def timeSteps(self_nd2_file):
    time_steps = []
    for i in range(self_nd2_file._rdr._seq_count()):
        coords = self_nd2_file._rdr._coords_from_seq_index(i)
        _chs = self_nd2_file._rdr._frame_metadata(i).get('channels')
        stamp = _chs[0]['time']['relativeTimeMs']
        time_steps.append(stamp/1000.0)
        # print(coords, stamp)
    return time_steps
```

```python
time_steps = timeSteps(nd2_file)
```

```python
print(f'num time steps: {len(time_steps)}')
print(f'{time_steps}')
```

```python
def timeStamp(self_nd2_file, frame_num):
    _chs = self_nd2_file._rdr._frame_metadata(frame_num).get('channels')
    time_stamp = _chs[0]['time']['relativeTimeMs']
    return time_stamp

def getTimeStamps(self_nd2_file):
    time_stamps = []
    max_frame_num = self_nd2_file._rdr._seq_count() # is this different to the num of frames from the xarray? probably not but worth checking?
    for frame_num in range(max_frame_num):
        time_stamps.append(timeStamp(self_nd2_file, frame_num)/1000.0)
    return time_stamps
```
